```python
# import the dataset from the local repository
import pandas as pd
df = pd.read_csv('auto_theft.csv')
df
```

```python
# fill null values with 0
df.fillna(0, inplace = True)
df
```

```python
# change the datatype into int
df['Auto_Theft_Coordinated/Traced'] = df['Auto_Theft_Coordinated/Traced'].astype(int)
df['Auto_Theft_Recovered'] = df['Auto_Theft_Recovered'].astype(int)
df
```

```python
# drop unneccesary columns from dataframe
autotheft = df.drop(['Group_Name', 'Sub_Group_Name'], axis = 1)
autotheft
```

```python
import matplotlib.pyplot as plt
def line_plot(autotheft):
    areas = list(autotheft.iloc[:, 0].unique())
    time = list(autotheft.iloc[:, 1].unique())
    for i in time:
        stolen = []
        for j in areas:
            new = autotheft[autotheft["Year"] == i]
            stolen.append(sum(new[new["Area_Name"] == j].iloc[:, 4]))
        s = str(i) + ' Auto Theft Stolen District Wise'
        plt.figure(figsize = (12, 8))
        plt.bar(areas, stolen, 0.25)
        plt.title(s,  bbox={'facecolor':'0.8', 'pad':5}, loc = 'center')
        plt.ylabel('No. of Autovehicle Stolen')
        plt.xticks(rotation = 90)
        plt.show()
        stolen.clear()
```

```python
line_plot(autotheft)
```
