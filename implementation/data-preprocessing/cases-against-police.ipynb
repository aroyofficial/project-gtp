```python
# select neccessary columns from 2016 Cases against Police Personnels.csv dataset
import pandas as pd
df1 = pd.read_csv('2016 Cases against Police Personnels.csv', warn_bad_lines = False, error_bad_lines = False)
df1 = df1.iloc[:, [2, 3, 4, 5, 6]]
df1['Year'] = [1 for i in range(len(df1))]
df1.rename(columns={'Total Number of Criminal Cases Registered during the year': 'Total Cases', 'Police Personnel Arrested': 'Police Arrested', 'Cases Chargesheeted': 'Chargesheets', 'Police Personnel Chargesheeted': 'Chargesheet against police'}, inplace = True)
df1
```

```python
# select neccessary columns from 2017 Cases against Police Personnels.csv dataset
df2 = pd.read_csv('2017 Cases against Police Personnels.csv', warn_bad_lines = False, error_bad_lines = False)
df2 = df2.iloc[:, [2, 3, 7, 5, 8]]
df2['Year'] = [2 for i in range(len(df2))]
df2.rename(columns={'Number of Cases - Registered': 'Total Cases', 'Number of Police Personel - Arrested': 'Police Arrested', 'Number of Cases - Chargesheeted': 'Chargesheets', 'Number of Police Personel - Chargesheeted': 'Chargesheet against police'}, inplace = True)
df2
```

```python
# select neccessary columns from 2018 Cases against Police Personnels.csv dataset
df3 = pd.read_csv('2018 Cases against Police Personnels.csv', warn_bad_lines = False, error_bad_lines = False)
df3 = df3.iloc[:, [2, 3, 7, 5, 8]]
df3['Year'] = [3 for i in range(len(df3))]
df3.rename(columns={'Number of Cases - Registered': 'Total Cases', 'Number of Police Personel - Arrested': 'Police Arrested', 'Number of Cases - Chargesheeted': 'Chargesheets', 'Number of Police Personel - Chargesheeted': 'Chargesheet against police'}, inplace = True)
df3
```

```python
# create a new dataframe for cases against police personnels
against_police = pd.concat([df1, df2, df3])
against_police
```

```python
# select neccessary columns from 2016 Escapes from Police Custody.csv dataset
df1 = pd.read_csv('2016 Escapes from Police Custody.csv', warn_bad_lines = False, error_bad_lines = False)
df1 = df1.iloc[:, [2, 4, 7, 8]]
data = []
for i in range(len(df1)):
    data.append(df1.iloc[i, 2] + df1.iloc[i, 3])
df1['Rearrested'] = data
del df1['Escapees Rearrested who Escaped - From Lockup']
del df1['Escapees Rearrested who Escaped - Other than Lockup']
df1['Year'] = [1 for i in range(len(df1))]
df1.rename(columns={'Total Number of Persons Escaped from Police Custody': 'Escaped'}, inplace = True)
df1
```

```python
# select neccessary columns from 2017 Escapes from Police Custody.csv dataset
df2 = pd.read_csv('2017 Escapes from Police Custody.csv', warn_bad_lines = False, error_bad_lines = False)
df2 = df2.iloc[:, [2, 6, 7]]
df2['Year'] = [2 for i in range(len(df2))]
df2.rename(columns={'Persons Escaped - Total': 'Escaped', 'Escapees - Re-Arrested': 'Rearrested'}, inplace = True)
df2
```

```python
# select neccessary columns from 2018 Escapes from Police Custody.csv dataset
df3 = pd.read_csv('2018 Escapes from Police Custody.csv', warn_bad_lines = False, error_bad_lines = False)
df3 = df3.iloc[:, [2, 6, 7]]
df3['Year'] = [3 for i in range(len(df3))]
df3.rename(columns={'Persons Escaped - Total': 'Escaped', 'Escapees - Re-Arrested': 'Rearrested'}, inplace = True)
df3
```

```python
# create a new dataframe for jail escapees
jail_escape = pd.concat([df1, df2, df3])
jail_escape
```

```python
# now concat two, against_police and jail_escape dataframe
jail_crimes = pd.concat([against_police, jail_escape], axis=1, join='outer')
jail_crimes.drop(jail_crimes.columns[[6, 9]], axis = 1, inplace = True)
jail_crimes
```
